```python
import time
import random


def interpolation_search(arr, target):
    """
    Interpolation Search
    Returns: (index, comparisons)
    """
    low = 0
    high = len(arr) - 1
    comparisons = 0

    while (
        low <= high
        and arr[low] <= target <= arr[high]
    ):
        comparisons += 1

        # If only one element remains
        if low == high:
            if arr[low] == target:
                return low, comparisons
            return -1, comparisons

        # Prevent division by zero
        if arr[high] == arr[low]:
            break

        # Interpolation formula
        pos = low + int(
            ((target - arr[low]) * (high - low))
            / (arr[high] - arr[low])
        )

        # Ensure position stays within bounds
        if pos < low or pos > high:
            break

        if arr[pos] == target:
            return pos, comparisons
        elif arr[pos] < target:
            low = pos + 1
        else:
            high = pos - 1

    return -1, comparisons


def binary_search(arr, target):
    """
    Binary Search
    Returns: (index, comparisons)
    """
    low = 0
    high = len(arr) - 1
    comparisons = 0

    while low <= high:
        comparisons += 1
        mid = (low + high) // 2

        if arr[mid] == target:
            return mid, comparisons
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1

    return -1, comparisons


def performance_analysis():
    sizes = [1000, 5000, 10000, 50000, 100000]

    print(
        f"{'Size':>10} {'IS Time(ms)':>14} {'BS Time(ms)':>14} "
        f"{'IS Comparisons':>18} {'BS Comparisons':>18}"
    )
    print("-" * 80)

    for size in sizes:
        # Create a sorted random array
        arr = sorted(random.sample(range(size * 10), size))

        # Random target from the array
        target = random.choice(arr)

        # -------- Interpolation Search --------
        start = time.perf_counter()

        for _ in range(100):
            idx_is, comp_is = interpolation_search(arr, target)

        is_time = ((time.perf_counter() - start) / 100) * 1000

        # -------- Binary Search --------
        start = time.perf_counter()

        for _ in range(100):
            idx_bs, comp_bs = binary_search(arr, target)

        bs_time = ((time.perf_counter() - start) / 100) * 1000

        print(
            f"{size:>10} "
            f"{is_time:>14.4f} "
            f"{bs_time:>14.4f} "
            f"{comp_is:>18} "
            f"{comp_bs:>18}"
        )


def main():
    arr = [2, 5, 10, 15, 23, 35, 48, 60, 75, 90, 105, 120]
    target = 35

    idx, comps = interpolation_search(arr, target)

    print("Interpolation Search Example")
    print("-" * 35)
    print("Array :", arr)
    print("Target:", target)

    if idx != -1:
        print(f"Element found at index {idx}")
    else:
        print("Element not found")

    print(f"Comparisons: {comps}")

    print("\nPerformance Analysis")
    performance_analysis()


if __name__ == "__main__":
    main()
```
